```python
def calculate_press_time1(distance):
    a = 0.0021
    press_time = a * distance
    return press_time

def calculate_press_time2(distance):
    press_time = 0.0024 * distance - 0.06
    return press_time
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 400, 400)
y1 = calculate_press_time1(x)
y2 = calculate_press_time2(x)

plt.plot(x, y1, label='y = 0.0021 * x')
plt.plot(x, y2, label='y = 0.0024 * x - 0.06')

plt.xlabel('distance')
plt.ylabel('press time')

plt.legend()
plt.show()
```
